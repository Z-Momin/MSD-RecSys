```python
import sqlite3
import pandas as pd
import tsne as t
```

```python
dat = sqlite3.connect('/Users/farrisatif/Desktop/NYU DS/1004/track_metadata.db')
query = dat.execute("SELECT * From songs")
cols = [column[0] for column in query.description]
results= pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

```python
results1 = results.iloc[0:100,:]
results1
```

```python
out = t.tsne(results1[[ 'duration', 'artist_familiarity', 'artist_hotttnesss',
       'year', 'track_7digitalid', 'shs_perf', 'shs_work']].to_numpy(),\
       no_dims=2, initial_dims=7, perplexity=1000)
```
